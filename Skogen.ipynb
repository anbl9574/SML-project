```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_data = pd.read_csv('training_data.csv')
input_data
```

```python
input_data.describe()
```

```python
# Create a new column 'increase_stock_binary' and map the values to low_bike_demand = 0 and high_bike_demand = 1
input_data_binary = input_data.copy() 
input_data_binary['increase_stock_binary'] = input_data_binary['increase_stock'].map({'low_bike_demand': 0, 'high_bike_demand': 1})

#Dropping original increase stock since it is no longer needed, and droping snow since its allways =0
input_data_binary.drop('increase_stock', axis=1, inplace=True)
input_data_binary.drop('snow', axis=1, inplace=True)

input_data_binary
```

```python
# Labels are the values we want to predict
labels_2_predict = np.array(input_data_binary['increase_stock_binary'])

# Romve column we want to predict
input_data_drop_stock = input_data_binary.drop('increase_stock_binary', axis = 1)

# Saving column names for later use
input_data_drop_stock_list = list(input_data_drop_stock.columns)

# Convert to numpy array
input_data_drop_stock = np.array(input_data_drop_stock)

```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
training_data, test_data, training_predict, test_predict = train_test_split(input_data_drop_stock, labels_2_predict, test_size = 0.10)

# Import the model we are using
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import GridSearchCV

# Parameters for the model
"""
n_estimators: The number of trees in the forest.

criterion: The function used to measure the quality of a split. It can be either 'gini' for Gini impurity or 'entropy' for information gain.

max_depth: The maximum depth of the tree. If None, the nodes are expanded until all leaves are pure or until they contain less than the minimum samples split.

min_samples_split: The minimum number of samples required to split an internal node.

min_samples_leaf: The minimum number of samples required to be at a leaf node.

min_weight_fraction_leaf: The minimum weighted fraction of the sum total of weights (of all the input samples) required to be at a leaf node.

max_features: The number of features to consider when looking for the best split. It can be an int, float, 'auto', 'sqrt', 'log2', or None.

max_leaf_nodes: Grow trees with max_leaf_nodes in best-first fashion.

min_impurity_decrease: A node will be split if this split induces a decrease of the impurity greater than or equal to this value.

min_impurity_split: Threshold for early stopping in tree growth. A node will split if its impurity is above the threshold; otherwise, it is a leaf.

bootstrap: Whether bootstrap samples are used when building trees.

oob_score: Whether to use out-of-bag samples to estimate the generalization accuracy.

n_jobs: The number of jobs to run in parallel for both fit and predict. -1 means using all processors.

random_state: If int, random_state is the seed used by the random number generator; If RandomState instance, random_state is the random number generator; If None, the random number generator is the RandomState instance used by np.random.


"""
Froet = RandomForestClassifier()

Parameters = { 
    'n_estimators': [200, 1000],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy'],
    'n_jobs' : [-1]
}
Skogen = GridSearchCV(Froet, Parameters, scoring='accuracy', refit=True)
Skogen.fit(training_data, training_predict)

# Train the model on training data
#Skogen.fit(training_data, training_predict)

# Use model on test_data to check validity
model_prediction = Skogen.predict(test_data)

```

```python
best_params = Skogen.best_params_

print("Best Hyperparameters:", best_params)
```

```python
#Correct prediction by the modell as a procentage

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(test_predict, model_prediction)
print(f"Accuracy: {accuracy}")
```

```python
#Precision is a measurement of how correct the prediction is, if the model predicts 1 how likely is it that the correct answear is 1
#Precision is calculated with: True positive / (True positive + False positive)

#Recal is a measurement of how good the model is at captureing all positive instances, for example how good is the modell at finding all 0
#Recal is calcultaed with: True positive / (True positive + False Negative)

#F1 is a harmonic mean between precision and recal, that is usefull when the class distribution is very unewen such as in our case where there are alot more 0 than 1.
#F1 is calculated with: 2 * Precision * Recall / (Precision + Recall)

#Support is to show how many of each instance there is
#Accuracy is the correct predicion by the modell as a procentage
#Macro average, average over Precision, Recal, and F1
#Weighted average, average over Precision, Recal and F1 with weighted values takeing in to acount the amount of datapoints in each class (0 or 1)



from sklearn.metrics import classification_report
print("Classification Report:")
print(classification_report(test_predict, model_prediction))
```

```python
#Confusion matrix displayes amount of, True positiv (topleft), False positiv (bottom left), False negative (top right), True negative (bottom left)


from sklearn.metrics import confusion_matrix
cm = confusion_matrix(test_predict, model_prediction)
print("Confusion Matrix:")
print(cm)
```
